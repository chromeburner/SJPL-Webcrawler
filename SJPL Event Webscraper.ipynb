```python
import requests
url = 'https://sjpl.bibliocommons.com/v2/events?locations=25&endDate=2023-08-31'

#making the get request
response = requests.get(url)

#parsing the data
from bs4 import BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')

date = soup.findAll('span', 'cp-screen-reader-message')


for date in date:
    print(date.text)
```
